# Splines cúbicos

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()
```

# Tarea
* Grafique los splines cúbicos tal que cumplan con una pendiente de $m=-3$ en los puntos dados.

```python
import numpy as np, matplotlib.pyplot as plt

def S0(x): return 1 + 7.5*(x+1) - 3.5*(x+1)**3
def S1(x): return 5 - 3*x - 10.5*x**2 + 11.5*x**3

xs = [-1, 0, 1]; ys = [1, 5, 3]

x1 = np.linspace(-1, 0, 100)
x2 = np.linspace(0, 1, 100)

plt.plot(x1, S0(x1), 'r')
plt.plot(x2, S1(x2), 'r')
plt.scatter(xs, ys, color='k')
plt.axvline(0, ls='--', lw=0.5)
plt.title('Splines cúbicos con pendiente -3 en x=0')
plt.grid(); plt.show()
```
